Note that Email,Address and Avatar are of no use to us for module building.From above dataset we used  "Avg. Session Length",	"Time on App",	"Time on Website",	"Length of Membership" as a independent features and	"Yearly Amount Spent" as a dependent features so work on them only. 

```python
df.drop(['Email', 'Address','Avatar'], axis = 1, inplace = True)

```

Lets use "Avg. Session Length", "Time on App", "Time on Website", "Length of Membership" as a independent features and "Yearly Amount Spent" as a dependent features.

```python
X = df[['Avg. Session Length', 'Time on App', 'Time on Website','Length of Membership']]
y = df['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(X_train,y_train)
```

```python
X_train
```

```python
#We'll save the model to our local machine:
import pickle
pickle.dump(lr_model, open("lr_model_pkl","wb"))

```

```python
y_pred = lr_model.predict(X_test)
```

To evaluate our model performance, we will be calculating the residual sum of squares and the explained variance score (R2).

```python
X_test
```

```python
y_pred
```

```python
X_test
```

```python
from sklearn.metrics import r2_score
y_pred = lr_model.predict(X_test)
r2_score(y_test,y_pred)
```

```python
from sklearn import metrics
print('MAE :'," ", metrics.mean_absolute_error(y_test,y_pred ))
print('MSE :'," ", metrics.mean_squared_error(y_test,y_pred ))
print('RMAE :'," ", np.sqrt(metrics.mean_squared_error(y_test,y_pred )))
```

```python
lr_model.coef_
```

```python
coeffecients=pd.DataFrame(lr_model.coef_,X.columns,columns=['Coeff'] )
coeffecients
```

##**How can you interpret these coefficients?**

- Keeping all other features constant, **a one-unit increase in Avg. Session Length is associated with an increase of 26.543543 total dollars spent yearly .**
- By Keeping all other features constant, **a one-unit increase in Time on App is associated with an increase of 38.362194 total dollars spent yearly .**
- Keeping all other features constant, **a one-unit increase in Time on the Website is associated with an increase of 0.735588 total dollars spent yearly .**
- Also, Keeping all other features constant, **a one-unit increase in Length of Membership is associated with an increase of 62.083781 total dollars spent yearly .**
